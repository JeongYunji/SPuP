### 개요

* 중심 극한 정리 : 표본 평균을 표준화하면 표본 개수가 증가함에 따라 표준정규 분포로 수렴한다.
* 균등 분포(uniform distribution)로부터 추출된 표본 평균의 분포는 표본 개수 n이 증가함에 따라 정규분포와 비슷해져 가는 것을 시각적으로 보인다.

* 균등 분포를 따르는 확률 변수를 생성하는 법
  * numpy module을 이용
  * 예를 들어, 0과 1 사이의 임의의 값을 발생하는 uniform 확률 변수 10개를 생성하려면, 다음의 방법을 이용하여 생성 가능하다.
    * 반환되는 값은 numpy array로써 각종 계산에 편리하게 응용할 수 있다.
    * 표본의 개수를 변경하려면 size=10의 부분을 수정하여 개수를 조정한다.

```python
import numpy as np

np.random.uniform(low=0, high=1, size=10)
```

0과 1사이의 균등분포 $U(0,1)$은 다음의 성질을 지닌다.
* 평균 : 0.5
* 분산 : $\frac{1}{12}$
* 표준편차 : $\frac{1}{\sqrt{12}}$

#### 히스토그램
* matplotlib.pyplot의 hist() 함수를 이용하면 histogram을 편리하게 작성할 수 있다.
* bins = "auto"는 histogram의 계급을 자동으로 설정한다는 의미.
* normed = True의 의미는 histogram을 normalize하여 histogram의 넓이의 총 합이 1이 되도록 한다는 뜻
* 그 외의 사항은 https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.hist 를 참고하여 자신만의 histogram을 작성하여 보자.

```python
import matplotlib.pyplot as plt
plt.hist(np.random.uniform(low=0, high=1, size=100), bins="auto", normed=True, facecolor = "none", edgecolor="black")
```
